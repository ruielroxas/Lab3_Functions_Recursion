```python
# CELL 2: RECURSIVE SIGNAL SHUTDOWN
CONTROL_NUM = SEED_NUMBER
FAVORITE_ARTIST = "CUP OF JOE"

def authorization_logger(func):
    """Decorator to log start and completion of the shutdown sequence."""
    def wrapper(*args, **kwargs):
        print("Authorization Started")
        result = func(*args, **kwargs)
        print("Authorization Completed")
        return result
    return wrapper

@authorization_logger
def signal_shutdown(power):
    """
    Recursively powers down the system.
    Returns the total number of recursive calls made.
    """
    # Base Case: power reaches 0
    if power <= 0:
        print("System Offline: Power at 0")
        return 0
    
    # Recursive Behavior
    print(f"Current signal strength: {power}")
    return 1 + signal_shutdown(power - 1)

# Initialize values
# Logic: power = CONTROL_NUM + len(FAVORITE_ARTIST)
initial_power = CONTROL_NUM + len(FAVORITE_ARTIST)

# Execute
total_calls = signal_shutdown(initial_power)
print(f"\nTotal Recursive Calls: {total_calls}")
```
